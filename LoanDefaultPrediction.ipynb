```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

FILE = "W:\\Default_Fin.csv"
cancer_data = pd.read_csv(FILE)

X = cancer_data
X
```

```python
X_n = cancer_data.drop(columns=["Defaulted?"], axis='columns')
Y_n = cancer_data["Defaulted?"]
X_n
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_n,Y_n,test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)

score = accuracy_score(Y_test, predictions)
score
```

```python
import joblib

joblib.dump(model, 'LoanDefaultPrediction.joblib')
```
